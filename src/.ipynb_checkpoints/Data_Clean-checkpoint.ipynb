```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
import glob
from sklearn.preprocessing import MinMaxScaler
```

```python
brent=pd.read_excel('../data/RBRTEd.xls',sheet_name='Data 1', header=0, skiprows=2)
brent.columns=['Date','Oil_Price']
brent.head()
```

```python
shares=["RDSB.L","BP.L","CNE.L","PMO.L","STL.OL","FP.PA","REP.MC","ENGI.PA","SLB.PA"]
share_price_df=pd.DataFrame();
for company in shares:
    file=pd.read_csv('../data/'+company+'.csv')
    file=file[['Date','Adj Close','Close']]
    file=file.dropna()
    file.columns=['Date','AdjustedSharePrice', 'ClosePrice']
    file['Company']=company
    file['Date']=file['Date'].astype('datetime64[D]')
    file['Year']=file['Date'].dt.year
    scaler=MinMaxScaler()
    file['AdjustedSharePriceScaled']=scaler.fit_transform(file['AdjustedSharePrice'].to_frame())
    share_price_df=share_price_df.append(file)
    
    
    
share_price_df=pd.merge(share_price_df,brent, on ='Date')    
```

```python
share_price_df.head()
```

```python
plt.rcParams['figure.figsize']=15,10
plt.rcParams['font.size']=15
fig, ax = plt.subplots(1,1)
brent.plot(x='Date', y='Oil_Price', grid=True, ax=ax, color='blue')
plt.ylabel('Oil Price in Dollars($) per barrel')
plt.title('Europe Brent Spot Price FOB')
plt.xlabel('Year')
plt.savefig('../results/Brent_Price.png')
```

```python
shares=["RDSB.L","BP.L","CNE.L","PMO.L","STL.OL","FP.PA","REP.MC","ENGI.PA","SLB.PA"]
sns.set()
palette= sns.cubehelix_palette(18, start=2, rot=0, dark=0, light=.95, reverse=False)
for company in shares:
    figure=sns.pairplot(share_price_df[share_price_df['Company']==company],hue='Year',palette=palette, 
    vars=['AdjustedSharePrice','Year','Oil_Price'], height=4,markers="o",
             plot_kws=dict(s=50, edgecolor="b", linewidth=0))
    
    fig = figure.fig 
    fig.subplots_adjust(top=0.93, wspace=0.3)
    fig.suptitle(str(company),fontsize=14, fontweight='bold')
    
    
    plt.savefig('../results/'+str(company)+'.png')
```

```python
shares1=["RDSB.L","BP.L","CNE.L","PMO.L","STL.OL","FP.PA","REP.MC","ENGI.PA","SLB.PA"]

for companies in shares1:
    plt.rcParams['figure.figsize']=30,30
    #plt.rcParams['font.size']=15
    sns.set(font_scale=2)
    ax=sns.catplot(data=share_price_df[np.logical_and(share_price_df.Company==companies,share_price_df.Year>2012)],
               kind='violin', x='Year', y= 'AdjustedSharePrice', height=7, aspect=1.2, inner='quart')
    ax.fig.suptitle(str('Company = ')+str(companies))
    plt.savefig('../results/'+str(companies)+str('shareprice')+'.png'),
```

```python
sns.set(font_scale=1.5)
sns.lmplot(x='Oil_Price', y='AdjustedSharePriceScaled', data=share_price_df[share_price_df.Year>2012],col='Company', col_wrap=3,
          order=1, line_kws={'color':'blue'}, scatter_kws={'color':'black', 'alpha':0.2})
plt.savefig('../results/AdjustedSharePriceScaledVsOilPrice.png')
```

```python
sns.set(font_scale=1.5)
palette_lm=sns.cubehelix_palette(5, start=4, rot=0, dark=0, light=0.85, reverse=False)
sns.lmplot(x='Oil_Price', y='AdjustedSharePriceScaled',hue='Year', data=share_price_df[share_price_df.Year>2012],col='Company', 
           col_wrap=3,order=1,palette=palette_lm, ci=None).set(xlim=(min(brent.Oil_Price),max(brent.Oil_Price)),ylim=(0,1))
plt.savefig('../results/AdjustedSharePriceScaledVsOilPrice_lm.png')
```
